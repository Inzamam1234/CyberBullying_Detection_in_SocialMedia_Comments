```python
import pandas as pd

df = pd.read_csv("../data/raw/train.csv")

df.head()
```

```python
df.shape
```

```python
label_cols = [
    "toxic",
    "severe_toxic",
    "obscene",
    "threat",
    "insult",
    "identity_hate"
]
```

```python
#Count occurrences of each label
df[label_cols].sum()
```

```python
import matplotlib.pyplot as plt

df[label_cols].sum().plot(kind="bar")
plt.title("Toxic Comment Distribution")
plt.show()
```

```python
# Check text length
df["length"] = df["comment_text"].apply(len)
```

```python
plt.hist(df["length"], bins=50)
plt.title("Comment Length Distribution")
plt.show()
```

```python
#Check how many comments have multiple toxic labels
df["label_count"] = df[label_cols].sum(axis=1)
df["label_count"].value_counts()
```
